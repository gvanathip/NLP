# Import Cleaned Data

# pythai

```python
df['pythainlp'] = df['sentence_without_space(x_clean)'].apply(word_tokenize,keep_whitespace=False)
```

```python
def to_binary(lst):
    binary_str = ""
    for i in lst:
        binary = '1'
        for j in range(len(i)-1):
            binary = binary+'0'
        binary_str = binary_str+binary
    return(binary_str)
```

```python
df['pythainlp_binary(y)'] = df['pythainlp'].apply(to_binary)
```

```python
df.head()
```

```python
def str_len(lst):
    lst = lst.replace(" ", "")
    return(len(lst))
```

```python
df['pythainlp_binary_len'] = df['binary(y)'].apply(str_len)
```

```python
df = df[df['binary_len']==df['pythainlp_binary_len']]
```

```python
y = df['binary(y)'].tolist()
```

```python
def merge(lst):
    list_merge = ''
    for i in range(len(lst)):
        list_merge = list_merge+lst[i]
    return(list_merge)
```
